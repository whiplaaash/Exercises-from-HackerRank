# Exercise 52:
Check if a given string is a floating point number.

```python
def is_float(element: any) -> bool:
    #If you expect None to be passed:
    if element is None: 
        return False
    try:
        float(element)
        return True
    except ValueError:
        return False

T = ['+0.5', '7.0', '.0O9']

for i in T:
    k = ['+', '-', '.']
    p = len(i.split('.'))
    j = i[0]
    rules = [j in k or j.isnumeric(), p == 2, is_float(i)]
    if all(rules):
        print(True)
    else:
        print(False)
```

# Exercise 53:

Recurrence series:

Define a recurrence series (Fibonacci series) and, using the map() function, find cubes of the first $n$ terms of the series.

```python
cube = lambda x: pow(x, 3)

def fibonacci(n):
    arr = []
    for i in range(n):
        if i == 0:
            k = 0
        elif i == 1:
            k = 1
        else:
            k = arr[i-1] + arr[i-2]
        arr.append(k)
    return arr

n = 7
print(list(map(cube, fibonacci(n))))
```

# Exercise 56:
Find the first occurence of the repeating character in a given string.

```python
import re

def findRep(s):
    
    match = re.findall(r'(\w)\1+', s)
    
    if match != []:
        print(match[0])
    else:
        print(-1)
```

```python
s = '12333345675534588111'
d = 'drf'
findRep(d)
```

# Exercise 57:

Use 'reduce' (and 'Fraction') to, first, write elements of a list as fractions and then to calculate the product of them.

```python
from fractions import Fraction
from functools import reduce

# This function takes fractions (a/b) as arguments, which means we have to specify a numerator and a denominator.
# It is possible by using the module 'Fraction'.

def product(fracs):
    t = reduce(lambda x, y: x*y, fracs)
    #t = reduce(Fraction.__mul__, fracs)
    return t.numerator, t.denominator

f = [Fraction(5,2), Fraction(3,7)]
product(f)
```

```python
fr = [[1, 2], [3, 4], [10, 6]]

arr = []
# Convert the list of lists into fractions.
for i in fr:
    k = Fraction(*map(int, i))
    arr.append(k)
    
print(arr)

# Calculate the product of fractions.
print(*product(arr))
```

# Exercise 58:

Regex:

Find all the groups that follow the pattern: (consonant) + (at least 2 vowels) + (consonant) in a given string.

Some example to understand how '\number' works.

```python
import re

pattern = r'(\w)(\w)(\w)\3'
string = 'aa8709nnnw55rt6677'

a = re.findall(pattern, string)
print(a)
```

The actual solution:

```python
pat = r'[qwrtypsdfghjklzxcvbnmQWRTYPSDFGHJKLZXCVBNM][aeiouAEIOU]{2,}[qwrtypsdfghjklzxcvbnmQWRTYPSDFGHJKLZXCVBNM]'
pattern = re.compile(pat)

S = 'weerta-akighkluujY+Yrty+AAv'

a = re.findall(pattern, S)

for i in a:
    if a != []:
        print(i[1:-1])
    else:
        print(-1)
```

Another version:

```python
vowels = 'aeiouAEIOU'
consonants = 'qwrtypsdfghjklzxcvbnmQWRTYPSDFGHJKLZXCVBNM'

pattern = r"(?<=[{}])([{}]{{2,}})(?=[{}])".format(consonants, vowels, consonants)

S = 'weerta-akighkluujY+Yrty+AA-v'

a = re.findall(pattern, S)


if a:
    for i in a:
        print(i)
else:
    print(-1)
```

# Exercise 59:

Regex:

Find the start index and the stop index (using start() and end()) of all (overlaping!) occurences of the substring $k$ in the string $S$.

```python
import re

S = 'aaajkaa'
k = 'aa'

matches = list(re.finditer(rf'(?=({k}))', S))

if matches:
    for match in matches:
        print((match.start(1), match.end(1)-1))
else:
    print((-1,-1))
```

Another version with 'escape()'.

```python
import re

S = 'aaajkaa'
k = 'aa'


#matches = list(re.finditer(rf'(?=({k}))', S))

matches = list(re.finditer('(?={})'.format(re.escape(k)), S))

if matches:
    for match in matches:
        print((match.start(), match.end()+1))
else:
    print((-1,-1))

#print(matches)
```

# Exercise 60:

Regex:

Replace all ' && ' and ' || ' (note the spaces) of the given (from input) text by 'and' and 'or', respectively.

```python
import re 

N = int(input())

for line in range(N):
    line = input()
    
    pat1 = r'(?<= )&&(?= )'
    pat2 = r'(?<= )\|\|(?= )'

    line = re.sub(pat1, 'and', line)
    line = re.sub(pat2, 'or', line)
               
    print(line)
```
