# Exercise 61:

Regex:

Check if the given string is a valid Roman numeral. (It can be up to 3999.)

```python
regex_pattern = r"^M{0,3}(CM|CD|D?C{0,3})(XC|XL|L?X{0,3})(IX|IV|V?I{0,3})$"

string = 'CDXXI'

import re
print(str(bool(re.match(regex_pattern, string))))
```

# Exercise 62:

Regex:

Check if the given string is a valid phone number. The valid phone number should start with 7, 8, or 9 and consists of 10 digits in total.

```python
import re

numb = '4567349684'
pattern = r'^[7|8|9][0-9]{9}$'
is_number = re.search(pattern, numb)

if is_number:
    print('YES')
else:
    print('NO')
```

# Exercise 63:

Regex:

Find if the given text contains HEX colors and print them. The HEX colors should start with '#' and has 3 or 6 characters, which include digits and letters a-f and A-F.

```python
text = ''' here are the colors #BE9;
kfl and next vol # 909090, '''
```

```python
import re 

n = 2

for line in range(n):
    line = input()
    # The pattern contains also one extra symbol at the end to fit the style of HackerRank'.
    pattern = r'\#([a-fA-F0-9]{3}|[a-fA-F0-9]{6})[;,\)]{1}'
    col = re.findall(pattern, line)
    
    if col:
        [print('#' + i) for i in col]
```

# Exercise 65:

Parsing HTML code:

Given the piece of code print the comments (single- and multi-line) and the data.

```python
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    
    def handle_data(self, data):
        if data.strip():
            print(f">>> Data\n{data}")
        else:
            pass

    def handle_comment(self, data): 
        if '\n' in data:
            print(f">>> Multi-line Comment\n{data.lstrip()}")
        else:
            print(f">>> Single-line Comment\n{data.lstrip()}")

html = ""       
for i in range(int(input())):
    html += input().rstrip()
    html += '\n'
    
parser = MyHTMLParser()
parser.feed(html)
parser.close()
```

# Exercise 67:

XML:

Count all attributes in the given piece of xml code.

```python
import sys
import xml.etree.ElementTree as etree

def get_attr_number(node):
    count = len(node.attrib)
    for child in node:
        count += get_attr_number(child)
    return count

if __name__ == '__main__':
    # Note that the code can be also provided as input.
    #sys.stdin.readline()
    #xml = sys.stdin.read()
    xml = '''
    <feed xml:lang='en'>
    <title>HackerRank</title>
    <subtitle lang='en'>Programming challenges</subtitle>
    <link rel='alternate' type='text/html' href='http://hackerrank.com/'/>
    <updated>2013-12-25T12:00:00</updated>
</feed>
'''
    tree = etree.ElementTree(etree.fromstring(xml))
    root = tree.getroot()
    print(get_attr_number(root))
```
