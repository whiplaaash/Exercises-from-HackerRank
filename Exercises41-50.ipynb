# Exercise 41: 
Capitalize each world in a given string.

Method 1:

```python
h = "hello world fgt"

def solve(s):
    a = s.split(' ')
    b = []
    for i in range(len(a)):
        k = a[i].capitalize()
        b.append(k)
    c = ' '.join(b)
    return c

solve(h)
```

# Exercise 42: 

Collections - Counter:

Given the number of shoes in the store, and the tuples with a size and a price, calculate the total sum of the sold shoes. Use "Counter" from "collections".

```python
from collections import Counter

# List of shoe sizes in the store.
list_shoes = [2, 3, 4, 5, 6, 8, 7, 6, 5, 18]

# List of tuples, with (size, price) format.
k=[(6,55), (6,45), (6,55), (4,40), (18,60), (10,50)]

cc = Counter(list_shoes)
sum = 0
for i in k:
    for j in cc.keys():
        p = i[0]
        if p == j and cc[j] != 0:
            sum = sum + i[1]
            cc[j] = cc[j] - 1
            
print(sum)
```

# Exercise 43: 

Minion game:

Find who will win the game and print his final score. Given a string, count all points for all possible substrings starting with any consonant (Stuart's points), then count all points for all possible substrings starting with any vowel (Kevin's points). Print the winner and his total number of points.

1st version:

```python
def minion_game(string):

    vow = list('AEIOU')
    sumStuart = 0
    sumKevin = 0
    
    for idx, val in enumerate(string):
        k = len(string[idx:])
        if val in vow:
            sumKevin = sumKevin + k
        else:
            sumStuart = sumStuart + k
            
    if sumStuart > sumKevin:
        print("Stuart", sumStuart)
    elif sumStuart < sumKevin:
        print("Kevin", sumKevin)
    else:
        print("Draw")
```

```python
p='BANANA'
minion_game(p)
```

2nd version (with numpy arrays, which make the computations faster):

```python
import numpy as np

def minion_game(string):

    vow = list('AEIOU')
    sumStuart = 0
    sumKevin = 0

    st = np.array(list(string))
    
    for idx, val in enumerate(st):
        k = len(st[idx:])
        if val in vow:
            sumKevin = sumKevin + k
        else:
            sumStuart = sumStuart + k
            
    if sumStuart > sumKevin:
        print("Stuart", sumStuart)
    elif sumStuart < sumKevin:
        print("Kevin", sumKevin)
    else:
        print("Draw")
```

```python
d = 'BANANA'
minion_game(d)
```

3rd version (using itertools):

```python
import itertools

def minion_game(string):

    vow = list('AEIOU')

    sumStuart = 0
    sumKevin = 0

    # Create all possible substrings.
    substrings = [''.join(string[i:j]) for i, j in itertools.combinations(range(len(string) + 1), 2)]

    for k in substrings:
        if k[0] in vow:
            sumKevin = sumKevin + 1
        else:
            sumStuart = sumStuart + 1
            
    if sumStuart > sumKevin:
        print("Stuart", sumStuart)
    elif sumStuart < sumKevin:
        print("Kevin", sumKevin)
    else:
        print("Draw")
```

```python
s = 'BANANA'
minion_game(s)
```

# Exerciese 47:

Regex:

Find if a given expression is regex or not.

```python
import re

def solve(regex):
    try:
        if re.search(r'(\*|\+|\?){2,}', regex):
            return False
        re.compile(regex)
        return True
    except re.error:
        return False
```

```python
s = '.\*'
solve(s)
```

# Exercise 48:

Itertools - product:

Find the maximal value of the function $S=(x_1^2+x_2^2+x_3^2+x_4^2)\% M$, where each $x_i$ is taken from different sets. Find such a combination of $x_i$ that $S$ is max.

```python
import itertools as it

a = {1, 7, 3}
b = {4, 5, 3, 1}
c = {4, 7, 1, 9}
d = {4, 7}

t = [a, b, c, d]

arr = []
M = 1000

for k in list(it.product(*t)):
    h = 0
    for l in k:
        h += l**2
    h = h%M
    arr.append(h)

print(max(arr))
```
